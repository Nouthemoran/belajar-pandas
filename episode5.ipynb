```python
import numpy as np 
```

```python
np.random.seed(100)
```

```python
angka=[]
for i in range(0,100):
    angka.append(i)
```

```python
angka
```

```python
baru = pd.DataFrame(np.random.rand(100, 5),angka,['A','B','C','D','E'])
```

```python
baru.head()
```

```python
baru.plot.area() # hist itu kotak2, klo area itu gelombang
```

```python
baru2 = pd.DataFrame(np.random.rand(4, 5),[0,1,2,3],['A','B','C','D','E'])
```
