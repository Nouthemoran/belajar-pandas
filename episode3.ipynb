```python
# klo bentuknya cantik itu dataframe, klo berantakan series
```

```python
dataku.iloc[-5:] # iloc ini punya 2 args yaitu [rows,column] 
# klo mau cari semua data pake ini : , klo minus contoh -5, itu artinya 5 trkhir
# klo positif itu yg awal awal
# : ini tuh kaya kata ( sampai ) 
```

```python
dataku.iloc[5:  ] # ini klo args ke 2 "-", itu bakal ngurangin data/rows
# terakhir yg di tampilin
```

```python
dataku.iloc[-5: -1, -4:-2] # klo di koma baru jdi column, : sampai yg -2 
# itu sampai berapa, yg -1 juga gitu.
```

```python
# slashing index
```

```python
dataku.iloc[[0,2,4]] # ini buat dapetin beberapa index
```

```python
dataku.iloc[[0,2,4] ,[1,3,6]] # bisa dapetin/iloc custom rows dan column
```

```python
# loc 
# .loc[rows,column]
```

```python
data_baru = pd.read_csv('kapal_titanic.csv', index_col='embarked') # jdi str
# stringggg bukan index lagi
```

```python
data_baru
```

```python
data_baru.loc['S']
```

```python
data_baru.loc['S','age'] # loc tdk harus berupa list jdi gausah dua squarebracket
```

```python
data_baru.loc['S',['age','fare']] # mengambil loc S dengan column yg di pilih
```

```python
data_baru.loc[['S','Q'],['age','fare']] # jdi kesimpulan aku kalo dua [[]], 
# itu tuh kalo buat beberapa data tpi buat argumen rows dan colum
```
